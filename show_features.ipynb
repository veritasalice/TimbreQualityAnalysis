```python
import matplotlib.pyplot as plt
import numpy as np
# Show an example
name = '1101_P1_oE0_1'
sample = np.load('../clarinetData/feature/{}.npy'.format(name)).T
sample_pu = np.load('../clarinetData/feature/{}_pu.npy'.format(name)).T
sample_pd = np.load('../clarinetData/feature/{}_pd.npy'.format(name)).T
sample_lu = np.load('../clarinetData/feature/{}_lu.npy'.format(name)).T
sample_ld = np.load('../clarinetData/feature/{}_ld.npy'.format(name)).T

```

```python
plt.figure(figsize=(18,12))
plt.subplot(231)
plt.imshow(sample_pu)
plt.title("pitch up")
plt.subplot(232)
plt.imshow(sample)
plt.title("original source")
plt.subplot(233)
plt.imshow(sample_pd)
plt.title("pitch down")

plt.subplot(234)
plt.imshow(sample_lu)
plt.title("loudness up")
plt.subplot(235)
plt.imshow(sample)
plt.title("original source")
plt.subplot(236)
plt.imshow(sample_ld)
plt.title("loudness down")
plt.show()
```
